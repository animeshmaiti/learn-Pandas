## Melt function
- Melt function is used to transform or reshape data.
- It is used to transform wide data to long data.
- It is used to transform long data to wide data.
- It is used to transform data from one format to another format.

```python
df1=pd.melt(df,id_vars=['day'],var_name='city')
# filter rows based on column value 
df1[df1['city']=='chicago']
```
