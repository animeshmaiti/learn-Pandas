```python
import pandas as pd
import numpy as np
df = pd.read_csv('weather_data-2.csv')
print(type(df['temperature'][0]))
df
```

### Replace with NaN for all values

```python
new_df = df.replace(-99999,np.NaN)
new_df
```

### Replace with NaN for particular columns with values

```python
new_df=df.replace({
    'temperature':-99999,
    'windspeed': [-99999,-88888],
    'event':'0'
},np.NaN)
new_df
```

### Replace with particular value for all values

```python
new_df=df.replace({
    -99999:np.NaN,
    -88888:np.NaN,
    '0':'Sunny'
})
new_df
```

```python
df2=pd.read_csv('weather_data-3.csv')
df2
```

### Using REGEX to replace values

```python
# applying in particular columns
new_df=df2.replace({
    'temperature':'[A-Za-z]',
    'windspeed':'[A-Za-z]'
},'',regex=True)
print(type(new_df['temperature'][0])) # temperature is still in string
new_df
```

```python
new_df['temperature']=new_df['temperature'].astype(int)
new_df['windspeed']=new_df['windspeed'].astype(int)
print(type(new_df['temperature'][0])) # temperature is now in int
new_df
```

```python
new_df=new_df.replace({
    -99999:np.NaN,
    -88888:np.NaN,
    '0':'Sunny'
})
new_df
```

### Now let's see how to replace all values in a dataframe with Values
Sunny=1<br>
Rain=2<br>
Snow=3

```python
new_df=new_df.replace(['Sunny','Rain','Snow'],[1,2,3])
new_df
```
