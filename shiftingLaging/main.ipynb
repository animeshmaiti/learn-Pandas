```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('fb.csv', parse_dates=['Date'], index_col='Date')
df
```

Shifting of data frame

```python
import matplotlib.pyplot as plt

# Plot original price
plt.plot(df.index, df['Price'], label='Original Price', marker='o')

# Plot shifted price for comparison
plt.plot(df.index, df['Price'].shift(1), label='Shifted Price', marker='o')

plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Comparison of Facebook Price with Shifted Price')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python
df['Prev day Price'] = df['Price'].shift(1)
df
```

```python
df['1 day Change'] = df['Price'] - df['Prev day Price']
df
```

```python
df['5 day % return'] = (df['Price'] - df['Price'].shift(5)) * 100 / df['Price'].shift(5)
df
```

```python
df = df[['Price']]
df
```
