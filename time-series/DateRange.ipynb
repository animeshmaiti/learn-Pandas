```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('appl2.csv')
df.shape
```

date_range() is a function that takes a list of strings, each string representing a date in the format "YYYY-MM-DD". The function returns the range of dates in the list as a string in the format "YYYY-MM-DD to YYYY-MM-DD".

```python
rng=pd.date_range(start='3/1/2021',end='3/30/2021',freq='B')
rng
```

```python
df.set_index(rng,inplace=True)
df.head()
```
