## Apple finance time series analysis

```python
import pandas as pd
```

```python
df=pd.read_csv('aapl.csv',parse_dates=["Date"],index_col='Date')
print(type(df.index[0]))
df.head()
```

```python
df['2021-01-01':'2021-01-07']
```

Give you the monthly data of apple stock price of 2021, our data is on daily basis, we converted monthly data by taking the average of the daily data.

```python
monthly_price=df.Close.resample('M').mean()
weakly_price=df.Close.resample('W').mean()
daily_price=df.Close
monthly_price
```

plot this monthly stock price

```python
%matplotlib inline
# monthly_price.plot()
weakly_price.plot()
```

```python
daily_price.plot()
```
